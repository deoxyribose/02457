## Densities and histograms

A probability density function $p(x)$ specifies that the probability of the
variable $x$ lying in the interval between any two points $a,b$ is
\begin{equation*}
P(x \in [a,b]) = \int_{a}^{b} p(x) dx
\end{equation*}
If $\{x_1,x_2,...,x_N\}$ is a set of points, the histogram of this set,
 evaluated on the ordered set of points $[z_1,z_2,...,z_M]$ is defined
 \begin{equation}
H_j =  \sum_{x_{k}  \in [z_{j},z_{j+1}]} 1 \ \ ,j=1,...,M-1
\end{equation}
and the normalized histogram is given by
\begin{equation}
\tilde{H}_j = \frac{H_j}{\sum_{j'=1}^{M-1} H_{j'}}
\end{equation}
The normalized histogram can be compared with the histogram approximation
to the density
\begin{equation}
P_j =  \int_{z_{j}}^{z_{j+1}} p(x) dx \ \ j=1,...,(M-1).
\end{equation}

We here focus on the density of the normal distribution:
\begin{equation}
p(x|\mu, \sigma^2) = \frac{1}{\sqrt{2\pi \sigma^2}}\exp \left( -\frac{(x-\mu)^2}{2\sigma^2}\right)
\end{equation}

```python
import numpy as np
import matplotlib.pylab as plt
% matplotlib inline
plt.rcParams['figure.figsize'] = (20,5)
```

```python
mu = 0     # True mean
sigma2 = 1 # True variance
xmin = -5  # min x value
xmax = 5   # max x value
Npdf = 100 # number of points in density
```

Gaussian densities can be written as zero-mean, unit-variance Gaussian (or standard Gaussian), scaled by the standard deviation, and translated by the mean
\begin{align}
\mathcal{N}(\mu,\sigma^2) = \sigma\mathcal{N}(0,1)+\mu
\end{align}
where
\begin{align}
\mathcal{N}(0,1) = \frac{1}{\sqrt{2\pi}}\exp{-\frac{1}{2}x^2}
\end{align}

```python
standard_gaussian_density = lambda x: np.exp(-.5*x**2)/np.sqrt(2*np.pi)
gaussian_density = lambda x,mu,sigma2: np.sqrt(sigma2)*standard_gaussian_density(x-mu)
```

```python
# Density
x1 = np.linspace(xmin,xmax,Npdf)
p1 = gaussian_density(x1,mu,sigma2)

# Ordered set of points for the histogram approximation to density
dx = .4 # bin width for histogram
z = np.linspace(xmin,xmax,(xmax-xmin)/dx)
p2 = dx*gaussian_density(z,mu,sigma2)
```

```python
plt.suptitle('Figure 1', size=20,y=1.04)
plt.subplot(121)
plt.plot(x1,p1)
plt.title('density of the 1D normal dist');
plt.axis('tight');
plt.ylim(0,.5);
plt.subplot(122)
plt.bar(z,p2,width=dx);
plt.title('histogram of the 1D normal dist');
plt.axis('tight');
plt.ylim(0,.5);
```

```python
# Now sample from a Gaussian
N = 100 # number of samples
y1 = np.sqrt(sigma2)*np.random.randn(N)+mu
y2,_ = np.histogram(y1,x2)
y2 = y2/y2.sum()
```

```python
plt.suptitle('Figure 2', size=20,y=1.04)
plt.subplot(121)
plt.bar(z[:-1],y2,width=dx,color='r')
plt.plot(z,dx*gaussian_density(z,mu,sigma2),'k--');
plt.title('histogram of samples and "true" histogram');
plt.axis('tight');
plt.ylim(0,.2);
plt.subplot(122)
plt.bar(z,p2,width=dx);
plt.title('"true" histogram');
plt.axis('tight');
plt.ylim(0,.2);
```

## Bayes' theorem

For real variables $x$ and labels $C_k=1,...,c$, Bayes' theorem reads,
\begin{equation}
P(C_k | x) = \frac{p(x|C_k) P(C_k)}{p(x)}
\end{equation}

If the class-conditional distributions are univariate normals with individual parameter sets
we can use the cells below to illustrate Bayes theorem.

### Checkpoint 1.2
Run the cells below that define three univariate normals and plot the resulting densities. Set
the *prior probabilities* $P(C_k)$ and plot the resulting $p(x)$ and
the posterior probabilities $P(C_k|x)$. Do this for different setting of the
prior probabilities and comment on the densities you get. Can you create a situation in which a class
has no points $x$ where it is most probable?

```python
xmin = -10
xmax = 10
Npdf = 1000 # number of points in grid
dx = .4

# Prior probabilities
P1, P2 = .4, .3
P3 = 1-P1-P2
```

```python
par1 = (-2,1) # mean -2, variance 1
par2 = (0,1)
par3 = (2,1)

x1 = np.linspace(xmin,xmax,Npdf)
p1 = gaussian_density(x1, *par1)
p2 = gaussian_density(x1, *par2)
p3 = gaussian_density(x1, *par3)
```

```python
px = p1*P1+p2*P2+p3*P3

P1X = p1*P1/px
P2X = p2*P2/px
P3X = p3*P3/px
```

```python
plt.suptitle('Figure 3',size=20,y=1.04)
plt.subplot(311)
plt.plot(x1,p1,'b',x1,p2,'r',x1,p3,'g');
plt.title('class conditional densities $p(x|c)$');
plt.subplot(312)
plt.plot(x1,px,'m');
plt.title('density $p(x)$');
plt.subplot(313)
plt.plot(x1,P1X,'b',x1,P2X,'r',x1,P3X,'g');
plt.title('posterior class densities $p(c|x)$');
plt.tight_layout()
```

## Decision boundaries

A decision rule,  is a division of the
space of $x$ so that each point is uniquely associated with
a class $C_k$. We can set up a simple 1D decision rule by dividing the real axis into three
intervals $I_1 = ]-\infty, d_1], I_2 = ]d_1, d_2], I_3 = ]d_2, \infty[$.

One way to summarize the errors of a decision rule is the
 error confusion matrix $R_{j,k}$  defined, e.g., as
\begin{equation}
R_{j,k} = \int_{I_j} p(x|C_k) dx
\end{equation}

### Checkpoint 1.3
Define the decision boundaries as above and compute

```python
# compute the confusion matrix for a given set of simple  decision boundaries
d12 = -3.8
d23 = max(1.0,d12)
plt.plot(x1,P1X,'b',x1,P2X,'r',x1,P3X,'g');
plt.axvline(d12,color='k');
plt.axvline(d23,color='k');
plt.title('posterior class probabilities P(c|x) and decision boundaries');
```

```python
from scipy.special import erf
```

```python
pars = np.array([par1,par2,par3])
```

```python
# error confusion matrix
R = np.zeros((3,3))
sqrt2 = np.sqrt(2)
R[0,:] = .5*(erf((d12-pars[:,0])/sqrt2/np.sqrt(pars[:,1]))+1)
R[1,:] = .5*(erf((d23-pars[:,0])/sqrt2/np.sqrt(pars[:,1]))-erf((d12-pars[:,0])/sqrt2/np.sqrt(pars[:,1])))
R[2,:] = .5*(1-erf((d23-pars[:,0])/sqrt2/np.sqrt(pars[:,1])))
```

```python
np.set_printoptions(suppress=True)
print("The error confusion matrix is:")
R
```

```python
print("Explain this alternative confusion matrix:")
CC = np.dot(np.diag([P1, P2, P3]),R)
CC
```
